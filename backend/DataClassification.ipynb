```
import pandas as pd
import numpy as np
import csv
import codecs
from collections import defaultdict
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pickle
```

```
df_pivoted = pd.read_csv("./df_pivoted.csv", encoding ="ISO-8859-1")
```

```
df_pivoted = df_pivoted.drop(df_pivoted.columns[0], axis=1)
```

```
df_pivoted
```

```
cols = df_pivoted.columns[1:]
```

```
cols
```

```
x = df_pivoted[cols]
y = df_pivoted['Source']
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

# Using Multinomial Classification

```
mnb = MultinomialNB()
mnb = mnb.fit(x_train, y_train)
```

```
mnb.score(x_test, y_test)
```

#Conclusion
As we have unique diseases with it's own symptoms, it doesn't make any sense to split the data and train a percent of them and test using the other data. we need to train all the data, all the diseases.

```
mnb_tot = MultinomialNB()
mnb_tot = mnb_tot.fit(x, y)
```

```
mnb_tot.score(x, y)
```

```
disease_pred = mnb_tot.predict(x)
```

```
disease_real = y.values
```

```
count = 0
for i in range(0, len(disease_real)):
    if disease_pred[i]==disease_real[i]:
        count+=1
        print ('Pred: {0} Actual:{1}'.format(disease_pred[i], disease_real[i]))
print(count)
```

```
count = 0
for i in range(0, len(disease_real)):
    if disease_pred[i]!=disease_real[i]:
        count+=1
        print ('Pred: {0} Actual:{1}'.format(disease_pred[i], disease_real[i]))
print(count)
```
